```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(69)
```

```python
S0 = 100.0   
K = 105.0    
r = 0.05     
sigma = 0.2  
T = 1.0      
steps = 252
dt = T / steps
N_paths = 100000 
```

```python

mu = r

Z = np.random.normal(size=(N_paths, steps))


increments = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
log_paths = np.cumsum(increments, axis=1)


log_ST = np.log(S0) + log_paths[:, -1]
ST = np.exp(log_ST)
ST[:10] 
```

```python

payoffs = np.maximum(ST - K, 0)
discounted = np.exp(-r * T) * payoffs


mc_price = discounted.mean()
mc_std = discounted.std(ddof=1)
mc_se = mc_std / np.sqrt(N_paths)
ci_low = mc_price - 1.96 * mc_se
ci_high = mc_price + 1.96 * mc_se

print(f"Plain Monte Carlo price estimate: {mc_price:.6f}")
print(f"Standard error: {mc_se:.6f}")
print(f"95% CI: [{ci_low:.6f}, {ci_high:.6f}]")
```

```python

N_half = N_paths // 2
Z_half = np.random.normal(size=(N_half, steps))

increments_half = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z_half
increments_anti = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * (-Z_half)

log_paths_half = np.cumsum(increments_half, axis=1)
log_paths_anti = np.cumsum(increments_anti, axis=1)

ST_half = np.exp(np.log(S0) + log_paths_half[:, -1])
ST_anti = np.exp(np.log(S0) + log_paths_anti[:, -1])

payoffs_pair = 0.5 * (np.maximum(ST_half - K, 0) + np.maximum(ST_anti - K, 0))
discounted_pair = np.exp(-r * T) * payoffs_pair

mc_price_anti = discounted_pair.mean()
mc_std_anti = discounted_pair.std(ddof=1)
mc_se_anti = mc_std_anti / np.sqrt(N_half)
ci_low_anti = mc_price_anti - 1.96 * mc_se_anti
ci_high_anti = mc_price_anti + 1.96 * mc_se_anti

print(f"Antithetic variates price estimate: {mc_price_anti:.6f}")
print(f"Standard error (antithetic): {mc_se_anti:.6f}")
print(f"95% CI (antithetic): [{ci_low_anti:.6f}, {ci_high_anti:.6f}]")

var_plain = mc_std**2
var_anti = mc_std_anti**2
reduction = 100.0 * (1 - var_anti / var_plain)
print(f"Variance reduction: {reduction:.2f}%")
```

```python

n_plot = 10
Z_plot = np.random.normal(size=(n_plot, steps))
increments_plot = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z_plot
log_paths_plot = np.cumsum(increments_plot, axis=1)
paths_plot = np.exp(np.log(S0) + log_paths_plot)

time_grid = np.linspace(dt, T, steps)

plt.figure(figsize=(10, 5))
for i in range(n_plot):
    plt.plot(time_grid, paths_plot[i])
plt.title("Sample GBM Paths (10 paths)")
plt.xlabel("Time (years)")
plt.ylabel("Stock price")
plt.grid(True)
plt.show()
```
